# News Reaction Model V9 Training Metrics

Plots four-class opportunity learning and calibration metrics.

```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
train_root = Path(r'D:/TradingML/runtimes/news-reaction-model/v9/train')
run_dir = max((p for p in train_root.iterdir() if (p / 'metrics.jsonl').exists()), key=lambda p: p.stat().st_mtime)
rows = [json.loads(line) for line in (run_dir / 'metrics.jsonl').read_text(encoding='utf-8').splitlines() if line.strip()]
frame = pd.json_normalize(rows)
display(frame.tail())
sample_column = next((n for n in ('sample', 'step', '_step') if n in frame), None)
columns = [n for n in ('train/loss', 'val/log_loss', 'val/accuracy', 'val/macro_f1', 'val/balanced_accuracy') if n in frame]
if sample_column and columns:
    frame.plot(x=sample_column, y=columns, subplots=True, figsize=(12, 10), grid=True)
    plt.tight_layout()
```
