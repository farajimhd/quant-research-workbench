# News Reaction Model v2 Diagram

Creates the default regression-only model and writes its architecture artifacts.

```python
from pathlib import Path
import torch
from research.mlops.model_artifacts import parameter_summary, write_model_artifacts
from research.news_reaction_model.v2.config import LoaderConfig, ModelConfig
from research.news_reaction_model.v2.data import make_dummy_batch
from research.news_reaction_model.v2.model import NewsReactionModelV2, build_model_mermaid

loader_config = LoaderConfig()
model_config = ModelConfig()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = NewsReactionModelV2(model_config).to(device)
print(parameter_summary(model))
write_model_artifacts(model=model, artifact_dir=Path('artifacts/news_reaction_model_v2'), model_config=model_config, input_contract={'embeddings': ['B', 2, 1024], 'chunk_mask': ['B', 2]}, output_contract={'return_forecasts': ['B', 10, 3]}, architecture_mermaid=build_model_mermaid(), summary_notes='Embedding-only regression model for actual abnormal terminal, high, and low returns.', dummy_input_factory=lambda: ((make_dummy_batch(2, loader_config, device=device).x,), {}))
```
