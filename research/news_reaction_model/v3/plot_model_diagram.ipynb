# News Reaction Model v3 Diagram

Creates the default embedding-only model and writes the same architecture artifacts as training.

```python
from pathlib import Path
import torch
from research.mlops.model_artifacts import parameter_summary, write_model_artifacts
from research.news_reaction_model.v3.config import LoaderConfig, ModelConfig
from research.news_reaction_model.v3.data import make_dummy_batch
from research.news_reaction_model.v3.model import NewsReactionModelV3, build_model_mermaid

loader_config = LoaderConfig()
model_config = ModelConfig()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = NewsReactionModelV3(model_config).to(device)
print(parameter_summary(model))
write_model_artifacts(model=model, artifact_dir=Path('artifacts/news_reaction_model_v3'), model_config=model_config, input_contract={'embeddings': ['B', 2, 1024], 'chunk_mask': ['B', 2]}, output_contract={'actionable_logits': ['B', 10, 2], 'direction_logits': ['B', 10, 2], 'magnitude_forecasts': ['B', 10, 3]}, architecture_mermaid=build_model_mermaid(), summary_notes='Hierarchical embedding-only news reaction model v3.', dummy_input_factory=lambda: ((make_dummy_batch(2, loader_config, device=device).x,), {}))
```
