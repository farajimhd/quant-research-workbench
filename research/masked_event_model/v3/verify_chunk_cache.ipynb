# Verify Chunk Cache

Checks sample event chunk parquet files before training.

```python
from pathlib import Path
import polars as pl
from research.masked_event_model.v3.config import DataConfig
from research.masked_event_model.v3.data import discover_chunk_files, target_horizons_from_columns

config = DataConfig(cache_root=Path(r'D:/market-data/flatfiles/us_stocks_sip/derived/event_chunks_v2'))
files = discover_chunk_files(config, start_date=config.train_start_date, end_date=config.test_end_date)
print('files', len(files))
for item in files[:10]:
    print(item.ticker, item.year_month, item.path)
```

```python
sample = files[0]
schema = pl.scan_parquet(str(sample.path)).collect_schema()
print('sample', sample.path)
print('horizons', target_horizons_from_columns(schema.names()))
print('columns', schema.names())
df = pl.scan_parquet(str(sample.path)).select(['ticker','session_date','chunk_start_ns','quote_count','trade_count','latest_mid']).head(20).collect()
df
```
