# Ticker Stream Block Inspection

Loads one direct ClickHouse ticker-stream block and prints event/origin/label shapes.

```python
from research.mlops.env import discover_env_files, load_env_files
from research.mlops.packed_market import ClickHouseTickerStreamConfig, ClickHouseTickerStreamDataset

load_env_files(discover_env_files(), verbose=True)
dataset = ClickHouseTickerStreamDataset(ClickHouseTickerStreamConfig(months=('2019-02',), tickers=('AAPL',), ticker_workers=1, max_blocks=1))
block = next(dataset.iter_blocks())
print('ticker', block.block_manifest.ticker)
print('month', block.block_manifest.month)
print('events', block.events.shape)
print('origins', block.origin_positions.shape)
print('labels', {k: v.shape for k, v in list(block.labels.items())[:8]})
print('first origin', int(block.origin_ordinals[0]), int(block.origin_timestamp_us[0]))
```
