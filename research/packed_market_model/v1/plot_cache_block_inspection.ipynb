# Packed Cache Block Inspection

Loads one packed block and prints event/origin/label shapes.

```python
from pathlib import Path
from research.mlops.packed_market import PackedMarketDataset, PackedMarketDatasetConfig

cache_root = Path(r'D:\\market-data\\prepared\\packed_market_block_cache\\packed_events_daily_index_2019-02')
dataset = PackedMarketDataset(PackedMarketDatasetConfig(cache_root=cache_root, months=('2019-02',), max_blocks=1))
block = next(dataset.iter_blocks())
print('ticker', block.block_manifest.ticker)
print('month', block.block_manifest.month)
print('events', block.events.shape)
print('origins', block.origin_positions.shape)
print('labels', {k: v.shape for k, v in block.labels.items()})
print('first origin', int(block.origin_ordinals[0]), int(block.origin_timestamp_us[0]))
```
